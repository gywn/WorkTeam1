```python
import numpy as np
import pandas as pd
import os.path

DSET_ROOT = "/Users/ever/Documents/HackathononKIBots/WorkTeam1/datasets/"
MAPPING_ROOT = "/Users/ever/Documents/HackathononKIBots/WorkTeam1/subtype_remapping/"

#DSET_ROOT = "https://raw.githubusercontent.com/gywn/WorkTeam1/master/datasets/all_nodes.json"

def load_est_data(est_data_path = DSET_ROOT):
    json_path = os.path.join(est_data_path,"all_nodes.json")
    return pd.read_json(json_path)

def load_type_mapping(ctype, est_data_path = MAPPING_ROOT):
    csv_path = os.path.join(est_data_path,ctype+"_subtypes.csv")
    dset = pd.read_csv(csv_path,encoding = "ISO-8859-1")
    dset.columns = ['a']
    dset=dset['a'].str.split(';', 1, expand=True)
    dset.columns = ['subtype','key']
    return dset


amenities_mapping = load_type_mapping('amenity')
tourism_mapping = load_type_mapping('tourism')
leisure_mapping = load_type_mapping('leisure')
shop_mapping = load_type_mapping('shop')


```
