```python
import numpy as np
import pandas as pd
import os.path

DSET_ROOT = "/Users/ever/Documents/HackathononKIBots/WorkTeam1/datasets/"

def load_est_data(est_data_path = DSET_ROOT):
    json_path=os.path.join(est_data_path,"all_nodes.json")
    return pd.read_json(json_path)

establishments = load_est_data()

```

```python
establishments['subtype'] = ""
establishments.head()
```

```python
tags = establishments.tags
types = establishments.type
str1 = types[32364]
str1[:7]
```

```python

for ii in range(4): #(establishments.shape[0]):
    type_str = types[ii]
    if type_str[:7] == "amenity":
        establishments.type[ii]="amenity"
        subtype_str = type_str[8:] 
        establishments.subtype[ii] = subtype_str

    elif type_str[:7] == "tourism":
        establishments.type[ii]="tourism"
        subtype_str = type_str[8:] 
        establishments.subtype[ii] = subtype_str

    else:
        establishments.type[ii]="leisure"
        subtype_str = type_str[8:] 
        establishments.subtype[ii] = subtype_str
```
